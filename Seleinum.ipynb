```python
#4
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

driver = webdriver.Chrome()
driver.get("https://www.flipkart.com/")
products=driver.find_element(By.CLASS_NAME,"Pke_EE")
products.send_keys('Sneakers')


products.send_keys(Keys.ENTER)
```

```python
brands=[]
product_description=[]
prices=[]

brand_tags=driver.find_elements(By.XPATH,'//div[@class="_2WkVRV"]')
for i in brand_tags:
    brand=i.text
    brands.append(brand)


description_tag = driver.find_elements(By.XPATH,'//div[@class="_2B099V"]/a[1]')
for i in description_tag:
    des = i.text.split('\n')[0]
    product_description.append(des)
    
price_tags=driver.find_elements(By.XPATH,'//div[@class="_3bPFwb"]')
for i in brand_tags:
    price=i.text
    prices.append(price)

df=pd.DataFrame({'Brand':brands,'Description':product_description,'Price':prices})
print(df)
```

```python
#5
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

driver = webdriver.Chrome()
driver.get("https://www.amazon.in/")
products=driver.find_element(By.ID,"twotabsearchtextbox")
products.send_keys('laptop')


products.send_keys(Keys.ENTER)
```

```python
titles=[]
ratings=[]
prices=[]

titles_tags=driver.find_elements(By.XPATH,'//div[@class="a-section a-spacing-none puis-padding-right-small s-title-instructions-style"]/h2')

for index,i in enumerate(titles_tags):
    if(index<=10):
        title=i.text
        titles.append(title)

#print(titles)

rating_tag = driver.find_elements(By.XPATH,'//div[@class="a-section a-spacing-small a-spacing-top-small"]/div[2]/div/span/span/a/i/span')
for i in rating_tag:
    rat = i.text
    ratings.append(rat)
    
#print(ratings)

price_tags=driver.find_elements(By.XPATH,'//div[@class="puisg-col puisg-col-4-of-12 puisg-col-4-of-16 puisg-col-4-of-20 puisg-col-4-of-24"]/div/div/div/div/a/span')
for index,i in enumerate(price_tags):
    if(index<=10):
        price=i.text
        prices.append(price)
#prices
df=pd.DataFrame({'title':titles,'Price':prices})
print(df)
```

```python
#6
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

driver = webdriver.Chrome()
driver.get("https://www.azquotes.com")
link=driver.find_element(By.XPATH,'//div[@class="mainmenu"]/ul/li[5]')
link.click()


```

```python
titles=[]
authors=[]
types= []
titles_tag=driver.find_elements(By.XPATH,'//div/p/a[@class="title"]')
for i in titles_tag:
    titles.append(i.text)
#print(titles)

author_tag=driver.find_elements(By.XPATH,'//div[@class="author"]/a')
for i in author_tag:
    authors.append(i.text)
    
types_tag=driver.find_elements(By.XPATH,'//div[@class="tags"]')
for i in types_tag:
    types.append(i.text)
    
    

#print(authors)

df=pd.DataFrame({'title':titles,'Author':authors,"Type":types})
print(df)
```
